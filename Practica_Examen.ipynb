1. Crear Dataframe con columnas "Producto, Precio Unitario, Cantidad ventas".

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = '/content/drive/MyDrive/IMLRN_Practicas/'
nombresillo = 'Examen.csv'
archivillo = open(path + nombresillo, 'w')
archivillo.close()
```

```python
df = pd.DataFrame(
    {
        "producto" : [],
        "precio unitario" : [],
        "cantidad ventas" : [],
    }
)

df.to_csv(path + "Examen.csv", index = False)
```

2. Agregar 10 productos preguntando al usuario.

```python
daf = pd.read_csv(path + 'Examen.csv')
```

```python
for i in range(10):
    producto = input(f"Ingrese el nombre del producto {i + 1}: ")
    precio_unitario = float(input(f"Ingrese el precio unitario del producto {i + 1}: "))
    cantidad_ventas = int(input(f"Ingrese la cantidad de ventas del producto {i + 1}: "))

    nueva_fila = {
        "producto": producto,
        "precio unitario": precio_unitario,
        "cantidad ventas": cantidad_ventas,
    }
    df = df.append(nueva_fila, ignore_index=True)
```

3. Agregar columna Sub Total (Precio Unitario * Cantidad Ventas).

```python
df['Sub Total'] = df['precio unitario'] * df['cantidad ventas']
```

4. Mostrar Dataframe con su información.

```python
df.info()
```

5. Mostrar los 3 productos más vendidos.

```python
df = df.sort_values(by='cantidad ventas', ascending = False)

productos_mas_vendidos = df.head(3)

print("Los 3 productos más vendidos:")
print(productos_mas_vendidos)
```

6. Imprimir la suma de la columna Sub Total.

```python
if 'Sub Total' not in df.columns:
    df['Sub Total'] = df['precio unitario'] * df['cantidad ventas']

suma_subtotal = df['Sub Total'].sum()

print(f"La suma de la columna 'Sub Total' es: {suma_subtotal}")
```

7. Imprimir el producto que tenga solo más de 10 unidades vendidas.

```python
productos_mas_de_10_unidades = df[df['cantidad ventas'] > 10]

print("Productos con más de 10 unidades vendidas:")
print(productos_mas_de_10_unidades[['producto', 'cantidad ventas']])
```

8. Mostrar los 3 productos con el Sub Total más bajo.

```python
df = df.sort_values(by='Sub Total', ascending=True)

productos_con_subtotal_mas_bajo = df.head(3)

print("Los 3 productos con el subtotal más bajo:")
print(productos_con_subtotal_mas_bajo)
```

9. Modificar 3 productos.

```python
nuevos_valores = [
    {"producto": "pantalon", "precio unitario": 15.99, "cantidad ventas": 20},
    {"producto": "blusa", "precio unitario": 12.49, "cantidad ventas": 15},
    {"producto": "chamarra", "precio unitario": 9.99, "cantidad ventas": 30},
]

for i, nuevo_valor in enumerate(nuevos_valores):
    df.loc[i * 2] = nuevo_valor

df.to_csv(path + nombresillo, index=False)

print("DataFrame actualizado:")
print(df)
```
