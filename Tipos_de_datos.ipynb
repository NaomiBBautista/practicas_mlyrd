```python
# Tipos de datos de secuencia
lis = [1, 2, 3]
tup = (1, 2, 3)
stri = 'numero'

print(type(lis))
print(type(tup))
print(type(stri))
```

```python
#Diccionario
dcc = {'nombre' : 'luis',
       'apellido' : 'torres',
       'edad' : 45 }

print(type(dcc))


#Boooleano
bol = True

print(type(bol))
```
