5. Crea un programa que calcule la factorial de un número ingresado por el usuario.

```python
def factorial(n):

   if n == 0 or n == 1:
            resultado = 1

   elif n > 1:
            resultado = n * factorial(n-1)

   return resultado

num_fac = int(input('Ingresa el número: '))
resultado = factorial(num_fac)
print('El resultado factorial del número es:', resultado)
```

6. Desarrolla  un  programa  que  genere  una  lista  de  números  primos  dentro  de  un  rango especificado por el usuario.

```python
def verificar_primo(nume):

    if nume <= 1:
        return False

    if nume <= 3:
        return True

    if nume % 2 == 0 or nume % 3 == 0:
        return False

    i = 5
    while i * i <= nume:

        if nume % i == 0 or nume % (i + 2) == 0:
            return False

        i += 6
    return True

def generar_primos(r1, r2):
    lista_primos = []

    for nume in range(max(2, r1), r2 + 1):

        if verificar_primo(nume):
            lista_primos.append(nume)

    return lista_primos

inicio = int(input("Ingrese el número inicial del rango: "))
fin = int(input("Ingrese el número final del rango: "))

lista_de_primos = generar_primos(inicio, fin)
print("Números primos en el rango:", lista_de_primos)
```

7. Desarrolla un programa que elimine elementos duplicados de una lista.

```python
def eliminar(lista):

    lista_sin_duplicados = []

    for elemento in lista:

        if elemento not in lista_sin_duplicados:
            lista_sin_duplicados.append(elemento)

    return lista_sin_duplicados

datos = input("Ingrese los elementos de la lista separados por coma: ")
lista_ingresada = datos.split(',')

lista_sin_duplicados = eliminar(lista_ingresada)

print("Lista sin duplicados:", lista_sin_duplicados)
```

8. Escribe un programa que encuentre el elemento máximo y mínimo en una lista de números (sin usar min ni max).

```python
def encontrar_numeros(lista):
    minimo = maximo = lista[0]

    for numero in lista:

        if numero < minimo:
            minimo = numero

        if numero > maximo:
            maximo = numero

    return minimo, maximo

lista_nume = [7,10,25,18,16,9,3,1,2,14,16]

minimo, maximo = encontrar_numeros(lista_nume)

print('De la siguiente lista de números: ', lista_nume)
print("El número mínimo es:", minimo)
print("El número máximo es:", maximo)
```

9. Realiza un programa que mezcle dos listas en una tercera lista.

```python
def mezclar_listas(l1, l2):
    lista_final = []

    for i in range(len(l1)):
        lista_final.append(l1[i])

    for i in range(len(l2)):
        lista_final.append(l2[i])

    return lista_final

lista_hombres = ['Andrés', 'Javier', 'Alejandro', 'Rogelio', 'Conde', 'Hector']
lista_mujeres = ['Naomi', 'Carla', 'Y ya, no hay mas']

lista_hm = mezclar_listas(lista_hombres, lista_mujeres)

print('Hombres en ISSC 311:', lista_hombres)
print('Mujeres en ISSC 311:', lista_mujeres)
print('Alumnos en ISSC 311:', lista_hm)
```

11. Desarrolla un programa que cuente cuántas veces aparece cada elemento en una lista.

```python
def contar_sabores(list_sabo):
    conteo = {}

    for num_sabo in list_sabo:

        if num_sabo in conteo:
            conteo[num_sabo] += 1

        else:
            conteo[num_sabo] = 1

    return conteo

lista_sabo = ['Fresa','Durazno','Mango','Chocolate','Vainilla','Chocolate', 'Fresa', 'Chocolate']
contear_sabo = contar_sabores(lista_sabo)
print('De la siguiente lista: \n', lista_sabo,'\n\nEl conteo de los sabores es:\n')

for elSabor, cantidad in contear_sabo.items():
    print(elSabor, 'aparece', cantidad, 'veces en la lista.')
```

15. Realiza un programa que indique si una contraseña ingresada por el usuario es fuerte o débil (basado en la longitud y la presencia de caracteres especiales).

```python
import re

def es_fuerte(contrasena):

    if len(contrasena) < 8:
        return False

    if not re.search(r'\d', contrasena):
        return False

    if not re.search(r'[A-Z]', contrasena):
        return False

    if not re.search(r'[a-z]', contrasena):
        return False

    if not re.search(r'[!@#$%^&*(),.?":{}|<>]', contrasena):
        return False

    return True

laContrasena = input('Ingrese una contraseña: ')

if es_fuerte(laContrasena):
    print('Felicidades la contraseña es fuerte.')
else:
    print('La contraseña es débil, debería crear una nueva')
```

27. Desarrolla una función que determine si un número es primo o no.

```python
def sera(elNu):
  if elNu <= 1:
    return False

  elif elNu <= 3:
    return True

  elif elNu % 2 == 0 or elNu % 3 == 0:
    return False

  i = 5
  while i * i <= elNu:
    if elNu % i == 0 or elNu % (i + 2) == 0:
      return False
      i += 6

  return True



elNume = int(input('Ingresa un número :'))

if sera(elNume):
  print(elNume, 'es un número primo.')

else:
  print(elNume,'no es un número primo, nimodillo.')
```

28. Escribe una función que convierta una lista de temperaturas de grados Celsius a Fahrenheit.

```python
def conver(list_cels):
  list_fahr = []

  for celsi in list_cels:
    fahr = (celsi * 9/5) + 32
    list_fahr.append(fahr)

  return list_fahr

tempC = [0, 10, 20, 30, 40]
tempF = conver(tempC)

print("Temperaturas en Celsius:", tempC)
print("Temperaturas en Fahrenheit:", tempF)
```

29. Crea una clase llamada Persona con atributos de nombre, edad y género, y funciones para imprimir y cambiar estos atributos.

```python
class Persona:
    def __init__(self, nombre, edad, genero):
        self.nombre = nombre
        self.edad = edad
        self.genero = genero

    def imprimir_informacion(self):
        print(f"Nombre: {self.nombre}")
        print(f"Edad: {self.edad}")
        print(f"Género: {self.genero}")

    def cambiar_nombre(self, nuevo_nombre):
        self.nombre = nuevo_nombre

    def cambiar_edad(self, nueva_edad):
        self.edad = nueva_edad

    def cambiar_genero(self, nuevo_genero):
        self.genero = nuevo_genero


personita1 = Persona("Alejandro", 21, "Masculino")
personita1.imprimir_informacion()

print('\nDatos modificados:')
personita1.cambiar_nombre("Alexito")
personita1.cambiar_edad(22)
personita1.cambiar_genero("Pansexual")
personita1.imprimir_informacion()
```

30. Crea una función que determine si una cadena de texto es un palíndromo.

```python
def es_palindromo(cadena):
    cadena = cadena.replace(" ", "").lower()

    return cadena == cadena[::-1]

cadenita = input('Ingrese una cadena de texto: ')

if es_palindromo(cadenita):
  print('La cadena ingresada es un palíndromo.')
else:
  print('La cadena ingresada no es un palíndromo.')
```
