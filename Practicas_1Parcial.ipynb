```python
# 5 : Serie Fibonacci
n = int(input('Ingrese la cantidad de números que desea imprimir: '))
lst = [1, 1]

print('\nNumero %i: %i' %(1, lst[0]))
print('Numero %i: %i' %(2, lst[1]))

# El .append sirve para agregar datos al final
for i in range(2, n):
  lst.append(lst[i - 1] + lst [i - 2])
  print('Número %i: %i' %(i + 1, lst[i]))
```

```python
# 10
def suma(n1, n2):
  total = n1 + n2
  return total

sum1 = suma(10, 16)
print('Ejemplo de función: %i' %(sum1))

n1 = int(input('\nIngresa el primer número: '))
n2 = int(input('Ingresa el segundo número: '))

sum2 = suma(n1, n2)
print('El resultado de la suma es: %i' %(sum2))
```

```python
# 11
def promedio (lista):

  return sum(lista) / len(lista)

numeros = input('Ingresa una lista de números separados por coma : ')
lista_num = [int(num) for num in numeros.split(",")]

resultado = promedio(lista_num)
print('El promedio de los números es: %i' %(resultado))
```

```python
# 12

def lon_palabras(*args):
  return [len(arg) for arg in args]

palabras = input('Ingresa una serie de palabras separadas por una coma: ')
lista_palabras = palabras.split(",")

longitud = lon_palabras(*lista_palabras)
print('La longitud de las palabras es: ', longitud)
```

```python
# 13
def factorial(n):
   if n == 0 or n == 1:
            resultado = 1
   elif n > 1:
            resultado = n * factorial(n-1)
   return resultado

num_fac = int(input('Ingresa el número: '))
resultado = factorial(num_fac)
print('El resultado factorial del número es:', resultado)
```
